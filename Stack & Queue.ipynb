#stack

```python
list=[]
while True:
    c=int(input('''
    1 push Elements
    2 pop Elements
    3 peek Element
    4 Display Elements
    5 Exit
    '''))
    if c==1:
        n=input("Enter The value ")
        list.append(n)
        print(list)
    elif c==2:
        if len(list)==0:
            print("Empty stack ")
        else:
            pop_element=list.pop()
            print(pop_element)
            print(list)
    elif c==3:
        if len(list)==0:
            print("Empty stack ")
        else: 
            print("last stack value ",list[-1])
    elif c==4:
        print("Display all stack values ",list)
    elif c==5:
        break
            
            
        
        
            
        
    
```

#Queue

```python
list=[]
while True:
    c=int(input('''
    1 Enqueue Elements
    2 Dequeue Elements
    3 Front Element
    4 rear Elements
    5 Display Queue
    6 Exit
    '''))
    if c==1:
        n=input("Enter The value ")
        list.append(n)
        print(list)
    elif c==2:
        if len(list)==0:
            print("Empty Queue ")
        else:
            del list[0]
        
            print(list)
    elif c==3:
        if len(list)==0:
            print("Empty Queue ")
        else: 
            print("first Queue value ",list[0])
    elif c==4:
        print("last all Queue values ",list[-1])
    elif c==5:
        print("Display Queue",list)
    elif c==6:
        break
    
            
            
        
        
            
        
    
```
